# Highrate

```python
data['CalculatedCharges']=data['MonthlyCharges']*df['tenure']
```

```python
data['TotalCharges']=pd.to_numeric(data['TotalCharges'],errors='coerce')
```

```python
data['TotalChargesStatus']=np.where(data['CalculatedCharges']<data['TotalCharges'],'Biggest','Smallest')
```

```python
data['TotalChargesStatus'].value_counts()
```

```python
data.head(5)
```

```python
df2 = data.nunique().reset_index()
colname = df2[df2[0]<=3]['index']
```

```python
for i in colname:
    print(i,":", data[i].unique())
```

```python
for j in "MultipleLines InternetService OnlineSecurity OnlineBackup DeviceProtection TechSupport StreamingTV StreamingMovies".split():
    data[j].replace("No internet service", "No", inplace=True)
```

```python
data.MultipleLines.replace("No phone service", "No", inplace=True)
```
